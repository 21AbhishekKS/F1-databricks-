```python
constructor_df = spark.table(f"{catalog_name}.{silver_schema}.constructors")
nationality_table = spark.table(f"{catalog_name}.{gold_schema}.ref_nationality_region")
```

```python
dim_constructor_df = constructor_df.join(
    nationality_table,
    constructor_df.nationality == nationality_table.nationality,
    "inner"
).select(
    constructor_df.constructor_id,
    constructor_df.constructor_name,
    constructor_df.nationality,
    nationality_table.region
)
```

```python
dim_constructor_df.write.mode("overwrite").saveAsTable(f"{catalog_name}.{gold_schema}.dim_constructor")
```
