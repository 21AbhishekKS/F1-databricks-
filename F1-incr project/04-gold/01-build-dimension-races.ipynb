```python
target_table = f"{catalog_name}.{gold_schema}.dim_races"
```

```python
races_table_path = f"{catalog_name}.{silver_schema}.races"
circuits_table_path = f"{catalog_name}.{silver_schema}.circuits"
```

```python
races_table = spark.table(races_table_path)
circuits_table = spark.table(circuits_table_path)
```

```python
from pyspark.sql import functions as F
```

```python
races_dim_df =(
     races_table.join(circuits_table,
                      races_table.circuit_id == circuits_table.circuit_id,
                      "inner"
                      ).select(races_table.season,
                               races_table.round,
                               races_table.race_name,
                               races_table.date,
                               circuits_table.circuitName,
                               circuits_table.locality,
                               circuits_table.country
                              )
)
```

```python
races_dim_df.write.mode("overwrite").saveAsTable(target_table)
```
