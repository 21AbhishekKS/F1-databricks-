```python
dbutils.widgets.text('batch_id' , "")
v_batch_id = dbutils.widgets.get('batch_id')
```

```python
source = f"{landing_file_path}/{v_batch_id}/results"
table_name = f"{catalog_name}.{bronze_schema}.results"
```

```python
schema = """
    constructorId STRING,
    date STRING,
    driverId STRING,
    grid LONG,
    laps LONG,
    number LONG,
    points DOUBLE,
    position LONG,
    positionText STRING,
    raceName STRING,
    round LONG,
    season LONG,
    status STRING,
    url STRING
"""
```

```python
results_df = (
    spark.read
    .format("json")
    #.option("inferSchema", True)
    .schema(schema)
    .load(source)
)

display(results_df)
```

```python
results_df = add_ingestion_metadata(results_df)
```

```python
(
    results_df.write
    .format("delta")
    .mode("overwrite")
    .saveAsTable(table_name)
    
)
```
