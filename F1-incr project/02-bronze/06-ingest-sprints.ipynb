```python
dbutils.widgets.text('batch_id' , "")
v_batch_id = dbutils.widgets.get('batch_id')
```

```python
source = f"{landing_file_path}/{v_batch_id}/sprints"
table_name = f"{catalog_name}.{bronze_schema}.sprints"
```

```python
sprints_schema = """
    constructorId STRING,
    date STRING,
    driverId STRING,
    grid LONG,
    laps LONG,
    number LONG,
    points DOUBLE,
    position LONG,
    positionText STRING,
    raceName STRING,
    round LONG,
    season LONG,
    status STRING,
    url STRING
"""
```

```python
sprints_df = (
    spark.read
    .format("json")
    .option("inferSchema", True)
    .schema(sprints_schema)
    .option('multiLine',True)
    .load(source)
)

display(sprints_df)
```

```python
sprints_df = add_ingestion_metadata(sprints_df)
```

```python
write_to_bronze(sprints_df,v_batch_id,table_name)

```
