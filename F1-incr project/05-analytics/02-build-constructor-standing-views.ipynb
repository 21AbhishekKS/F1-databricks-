```sql
CREATE OR REPLACE VIEW formula1.gold.v_constructor_standings
AS
WITH constructor_standings AS (
  SELECT 
    r.season,
    c.constructor_id,
    c.constructor_name,
    c.nationality,
    COUNT(*) as race_starts,
    SUM(r.points) as total_points,
    COUNT_IF(r.is_win) as number_of_wins,
    COUNT_IF(r.is_podium) as number_of_podium
  FROM 
    formula1.gold.fact_session_results r 
  JOIN
    formula1.gold.dim_constructor c 
    ON r.constructor_id = c.constructor_id
  GROUP BY 
    r.season,
    c.constructor_id,
    c.constructor_name,
    c.nationality
)
SELECT RANK() OVER(PARTITION BY season ORDER BY total_points DESC, number_of_wins DESC) as standings, *
FROM constructor_standings
ORDER BY season DESC, standings ASC;
```
