```python
drivers_df = spark.table(f"{catalog_name}.{silver_schema}.drivers")
nationality_table = spark.table(f"{catalog_name}.{gold_schema}.ref_nationality_region")
```

```python
dim_drivers_df = drivers_df.join(
    nationality_table,
    drivers_df.nationality == nationality_table.nationality,
    "inner"
).select(
    drivers_df.driver_id,
    drivers_df.date_of_birth,
    drivers_df.name,
    drivers_df.nationality,
    nationality_table.region
)
```

```python
dim_drivers_df.write.mode("overwrite").saveAsTable(f"{catalog_name}.{gold_schema}.dim_drivers")
```
