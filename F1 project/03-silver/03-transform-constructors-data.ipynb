```python
bronze_table = f"{catalog_name}.{bronze_schema}.constructors"
siver_table = f"{catalog_name}.{silver_schema}.constructors"
```

```python
constructors_df = spark.table(bronze_table)
display(constructors_df)
```

```python
required_constructor_df = constructors_df.select('constructorId', 'name', 'nationality', 'ingestion_timestamp' ,'source_file')
display(required_constructor_df)
```

```python
from pyspark.sql import functions as F
renamed_constructor_df = constructors_df.withColumnsRenamed(
    {
        "constructorId": "constructor_id",
        "name": "constructor_name",
        
    }
)
```

```python
titleCase_construcor_df = renamed_constructor_df.withColumn("nationality" ,  F.initcap(F.col("nationality")))

display(titleCase_construcor_df)
```

```python
distinct_constructor_df = titleCase_construcor_df.dropDuplicates(['constructor_id'])
display(distinct_constructor_df)
```

```python
(distinct_constructor_df.write
    .mode('overwrite')
    .saveAsTable(siver_table)
)
```
