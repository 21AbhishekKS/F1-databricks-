```python
source = f"{landing_file_path}/drivers.json"
table_name = f"{catalog_name}.{bronze_schema}.drivers"
```

```python
# Use STRUCT to define the nested name fields
schema = """
    dateOfBirth STRING,
    driverId STRING,
    name STRUCT<familyName: STRING, givenName: STRING>,
    nationality STRING,
    url STRING
"""
```

```python
drivers_df = (
    spark.read.
    format("json")
    #.option("inferSchema", True)
    .schema(schema)
    .load(source)
)

display(drivers_df)
```

```python
drivers_df = add_ingestion_metadata(drivers_df)
```

```python
(
    drivers_df.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable(table_name)
)
```
