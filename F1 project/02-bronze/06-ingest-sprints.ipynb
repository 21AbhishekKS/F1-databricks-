```python
source = f"{landing_file_path}/sprints"
table_name = f"{catalog_name}.{bronze_schema}.sprints"
```

```python
sprints_schema = """
    constructorId STRING,
    date STRING,
    driverId STRING,
    grid LONG,
    laps LONG,
    number LONG,
    points DOUBLE,
    position LONG,
    positionText STRING,
    raceName STRING,
    round LONG,
    season LONG,
    status STRING,
    url STRING
"""
```

```python
sprints_df = (
    spark.read
    .format("json")
    .option("inferSchema", True)
    .schema(sprints_schema)
    .option('multiLine',True)
    .load(source)
)

display(sprints_df)
```

```python
sprints_df = add_ingestion_metadata(sprints_df)
```

```python
(
    sprints_df.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .saveAsTable(table_name)
    
)

```
