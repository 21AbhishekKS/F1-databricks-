```sql
CREATE OR REPLACE VIEW formula1.gold.v_driver_standings
AS
WITH driver_standings AS (
  SELECT 
    r.season,
    d.driver_id,
    d.name,
    d.nationality,
    COUNT(*) as race_starts,
    SUM(r.points) as total_points,
    COUNT_IF(r.is_win) as number_of_wins,
    COUNT_IF(r.is_podium) as number_of_podium
  FROM 
    formula1.gold.fact_session_results r 
  JOIN
    formula1.gold.dim_drivers d 
    ON r.driver_id = d.driver_id
  GROUP BY 
    r.season,
    d.driver_id,
    d.name,
    d.nationality
)
SELECT RANK() OVER(PARTITION BY season ORDER BY total_points DESC, number_of_wins DESC) as standings, *
FROM driver_standings
ORDER BY season DESC, standings ASC;
```

```sql
select * from formula1.gold.v_driver_standings where season = 2025
```
